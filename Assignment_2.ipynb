```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv('House Price India.csv')
df.head()
```

```python
#Univariate Analysis
plt.hist(df['lot area'], bins=20)
plt.xlabel('lot area')
plt.ylabel('Density')
plt.title('Histogram of lot area')
plt.show()
```

```python
#Univariate Analysis
plt.hist(df['living area'], bins=20)
plt.xlabel('Living Area')
plt.ylabel('Density')
plt.title('Histogram of living area')
plt.show()
```

```python
#Univariate Analysis
plt.hist(df['Area of the house(excluding basement)'], bins=20)
plt.xlabel('House Area')
plt.ylabel('Density')
plt.title('Histogram of Area of the house')
plt.show()
```

```python
# Step 3: Bivariate Analysis
sns.scatterplot(x='lot area', y='living area', data=df)
plt.xlabel('lot area')
plt.ylabel('living area')
plt.title('Scatter plot between lot area and living area')
plt.show()
```

```python
# Step 3: Bivariate Analysis
sns.scatterplot(x='lot area', y='Price', data=df)
plt.xlabel('lot area')
plt.ylabel('Price')
plt.title('Scatter Plot between lot area and Price')
plt.show()
```

```python
# Step 3: Bivariate Analysis
sns.scatterplot(x='living area', y='Price', data=df)
plt.xlabel('living area')
plt.ylabel('Price')
plt.title('Scatter Plot between living area and Price')
plt.show()
```

```python
# Step 4: Multivariate Analysis
sns.pairplot(df[['living area', 'lot area', 'Price']])
plt.title('Pair Plot of living area, lot area and Price')
plt.show()
```

```python
# Step 4: Multivariate Analysis
sns.pairplot(df[['living_area_renov', 'lot_area_renov', 'Price']])
plt.title('Pair Plot of living area, lot area and Price after renovation')
plt.show()
```

```python
# Step 5: Descriptive Statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
# Step 6: Handle Missing Values
# Check for missing values
missing_values = df.isnull().sum()

# We impute missing values with the mean
df.fillna(df.mean(), inplace=True)

# After imputation, we recheck for missing values
missing_values_after_imputation = df.isnull().sum()

# Printing the summary of missing values before and after imputation
print("Missing Values Before Imputation:")
print(missing_values)

print("\nMissing Values After Imputation:")
print(missing_values_after_imputation)
```
